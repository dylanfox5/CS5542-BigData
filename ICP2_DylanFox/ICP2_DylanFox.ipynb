```python
from pyspark.sql import SparkSession
from pyspark import SparkConf, SparkContext

sc = SparkContext("local","ICP2")
```

```python
from google.colab import files
files.upload()
```

```python
import string

tf = sc.textFile("icp2 (5).txt")
tf_flat = tf.flatMap(lambda line: line.split(" ")).map(lambda line: line.translate(str.maketrans('', '', string.punctuation))).filter(lambda line: line)
tf_reduce = tf_flat.map(lambda word: (word.lower()[0], word + ",")).reduceByKey(lambda a,b: a+b);

tf_final = tf_reduce.map(lambda line: (line[0], set(line[1].split(","))))
```

```python
tf_final.saveAsTextFile("icp2_output_final.txt")
```
